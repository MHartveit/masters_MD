```python
import utils_clean
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
missing_pct = [0, 25,50,75, 90, 95]
```

```python
data = utils_clean.run("two_year_recid", "priors_count", "gender_factor", models = ["log_reg", "rf_cat", "knn"], dataset = "recid", n_runs = 4)
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.plot(missing_pct, data["Averaged results"]['mcar|acc|log_reg|fair_reg_50'], label = "fair_reg_50")
plt.plot(missing_pct, data["Averaged results"]['mcar|acc|log_reg|fair_reg_75'], label = "fair_reg_75")
plt.plot(missing_pct, data["Averaged results"]['mcar|acc|log_reg|fair_reg_99'], label = "fair_reg_99")
plt.plot(missing_pct, data["Averaged results"]['mcar|acc|log_reg|fair_reg_95'], label = "fair_reg_95")
plt.plot(missing_pct, data["Averaged results"]['mcar|acc|log_reg|mice_def'], label = "mice_def")
plt.plot(missing_pct, data["Averaged results"]['mcar|acc|log_reg|mean'], label = "mean")
plt.plot(missing_pct, data["Averaged results"]['mcar|acc|log_reg|cca'], label = "cca")
plt.legend()
plt.show()
```
